# *Analisando filmes**

```python
filmes = pd.read_csv("movies.csv")
filmes.head()
```

```python
notas.head()
```

```python
filmes.columns = ["filmeId", "titulo", "generos"]
filmes.head()
```

```python
notas.head()
```

```python
# O "query" é uma pergunta para a função
notas.query("filmeId==1")
```

```python
notas.query("filmeId==1").nota
```

```python
notas.query("filmeId==1").nota.mean()
# ToyStory (filme 1)
```

```python
notas.query("filmeId==2").nota.mean()
# Jumanji (Filme 2)
```

```python
notas.groupby("filmeId")
```

```python
#agrupei os filmes, agora vou tirar as médias de notas por filme
notas.groupby("filmeId").mean()
```

```python
#Apenas por notas do filme
medias_por_filme = notas.groupby("filmeId").mean()["nota"]
medias_por_filme.head()
```

```python
medias_por_filme.plot(kind='hist')
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(5,8))
sns.boxplot(x=medias_por_filme)
```

```python
medias_por_filme.describe()
```

```python
sns.displot(medias_por_filme)
```

```python
sns.displot(medias_por_filme, bins=10)
#bins são as caixinhas do gráfico
```

```python
import matplotlib.pyplot as plt
#biblioteca para fazer gráficos
plt.hist(medias_por_filme)
```

```python
plt.hist(medias_por_filme)
plt.title("Histograma das médias dos filmes")
```
